```python
from __future__ import print_function
import cv2
import matplotlib.pyplot as plt

%matplotlib inline

# Read the image from the matplotlib
IMAGE_FILE = './images/01-tyrex.jpg'
image = cv2.imread(IMAGE_FILE)

# Show the image with matplotlib
print("the originl image")
plt.imshow(image)
plt.show()
```

```python
# Access the pixel and manipulate it in the image
(b,g,r) = image[0,0] # get pixel from [0][0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r,g, b)) 

# Manipulating the image pixel
image[0,0] = (0,0,255)
(b,g,r) = image[0,0] # get pixel from [0][0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r,g, b)) 
```

```python
# check the image now, we won't know the difference, because it's one pixel
print("the manipulated image")
plt.imshow(image)
plt.show()
```

OpenCV stores RGB pixels in **reverse order**, so when we unpack and access each element in the tuple, we are actually viewing them in BGR order

### Getting and setting pixels like numpy 

```python
corner = image[0:100, 0:100] # get's square of 100X100 from top left corner
print("corner of 100X100 from top-left image")
plt.imshow(corner)
plt.show()
```

```python
# let's update all the pixels of the corner image
image[0:100, 0:100] = (0, 255, 0) # b, g, r --> puts green
print("corner of 100X100 from top-left image converted to green")
plt.imshow(image)
plt.show()
```

```python
# let's  convert all white of the image to green
import numpy as np

# Read the image from the matplotlib
IMAGE_FILE = './images/01-tyrex.jpg'
image = cv2.imread(IMAGE_FILE)

# Show the image with matplotlib
print("the originl image")
plt.imshow(image)
plt.show()

# convert white pixel of the images to green pixel
image[np.all(image == (255,255,255), axis=-1)] = (0, 250, 0)

# show the modified image
print("the modified image")
plt.imshow(image)
plt.show()
```
