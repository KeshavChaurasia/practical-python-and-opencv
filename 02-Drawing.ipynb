# Drawing

```python
# import package
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# prepare a canvas
canvas = np.zeros((300,300,3), dtype='uint8')

print("empty canvas")
plt.imshow(canvas)
plt.show()
```

```python
# let's draw
green = (0, 255, 0)
cv2.line(canvas, (0,0), (300,300), green) # canvas, start point, end point, color
print("green line")
plt.imshow(canvas)
plt.show()
```

```python
# let's draw
red = (255, 0, 0)
cv2.line(canvas, (300,0), (0,300), red, 2) # canvas, start point, end point, color, thickness
print("red line added")
plt.imshow(canvas)
plt.show()
```

```python
# rectangle
cv2.rectangle(canvas, (10,10), (60,60), green)
print("green rectangle added")
plt.imshow(canvas)
plt.show()
```

```python
cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)
print("red rectangle added")
plt.imshow(canvas)
plt.show()
```

```python
blue = (0,0,255)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1) # -1 is for filling
print("red rectangle added")
plt.imshow(canvas)
plt.show()
```

```python
# let's draw circles
```

```python
# fresh new canvas
canvas = np.zeros((300, 300, 3), dtype = "uint8")
```

```python
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2) # getting center of canvas
white = (255,255,255) 
```

```python
# draw circle of different radius
for r in range(0, 175, 25):
    cv2.circle(canvas, (centerX, centerY), r, white) # canvas, center, radius, color

print("drawing circle")
plt.imshow(canvas)
plt.show()
```

```python
# let's draw abstract circle
```

```python
plt.figure(figsize=(20,20))
plt.title("ABSTRACT IMAGES")
for j in range(1,5):
    plt.subplot(1,5,j+1)
    for i in range(1, 25):
        radius = np.random.randint(5, high = 200)
        color = np.random.randint(0, high = 256, size = (3,)).tolist()
        pt = np.random.randint(0, high = 300, size = (2,))
        cv2.circle(canvas, tuple(pt), radius, color, -1)
        
    plt.imshow(canvas)

plt.show()
```
